Phase 4 · Advanced NewsAPI options

```python
from langchain_config import newsapi
# Top headlines (faster, breaking news)
headlines = newsapi.get_top_headlines(q="NVIDIA", language='en', page_size=5)
for h in headlines['articles']:
    print(h['title'], "—", h['source']['name'])
```

```python
#  Filter by date range
from datetime import datetime, timedelta

week_ago = (datetime.now() - timedelta(days=7)).strftime('%Y-%m-%d')
today    = datetime.now().strftime('%Y-%m-%d')

articles = newsapi.get_everything(
    q="Fed interest rates",
    from_param=week_ago,
    to=today,
    language='en',
    sort_by='publishedAt'
)
for a in articles['articles'][:3]:
    print(a['publishedAt'][:10], "|", a['title'])
```
